### requests

get

```python
import requests as re
url = "https://book.naver.com/search/search.nhn"
params = {"query": "파이썬"}

response = re.get(url, params=params)
print(response.url)
```

```python
url = "https://www.naver.com"
headers = {"Content-Type": "application/json; charset=UTF-8"}

response = re.get(url, headers = headers)
print(response.url)
```

post

```python
url = "https://book.naver.com/search/search.nhn"
data = {"id": "ssodek"}

response = re.post(url, data=data)
print(response.url)
```

```python
url = "https://www.naver.com"
ress = re.get(url)
print(ress.encoding)
```

```python
url = "https://finance.naver.com"
ress = re.get(url)
print(ress.encoding)
print()
print(ress.text)
```

```python
url = "http://www.jjlim.net/me?id=kook&pass=okay"
print(upr.urlparse(url))
```
