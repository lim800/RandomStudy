```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# 편의점 시급 분석

```python
#데이터 호출
df = pd.read_csv('/Users/lesson6_mac/AtomData/python/판다스4/convenient_store.csv')
df.head()
```

```python
#데이터의 자료형
df.info()
```

```python
#데이터의 레코드수와 컬럼수 확인
df.shape
```

```python
#데이터의 요약 통계
df[['hourly_wage','outlier']].describe()
```

```python
#시급 요약 통계
df['hourly_wage'].describe()
```

```python
#지역 요약 통계
df['area'].describe()
```

```python
#시급 6천원 이상
df[df['hourly_wage']>=6000].head()
```

```python
#시급 기준으로 내림차순 정렬
df.sort_values(by=['hourly_wage'], ascending=False)
```

```python
#시급이 6000원 이상인 마포구에 위치한 편의점
df[(df['area1']=='마포구') & (df['hourly_wage'] >= 6000)]
```

```python
#CU 텍스트 항목 추출
df[df['name']=='CU'].head()
```

```python
#CU 텍스트가 포함된 PC방 추출
df[(df['name']=='CU') & (df['company']=='CUBE pc방')]
```

```python
#seoul 컬럼 추가 및 in seoul 문자열 할당
df['seoul'] = 'in Seoul'
df.head()
```

```python
#more_than_6000 컬럼 추가 및 시급이 6000원 이상이면 True 할당
df['more_than_6000'] = np.where(df['hourly_wage'] >=6000, 'True', 'False')
df.head()
```

```python
#6000원 이상 데이터의 요약 통계
df[df['hourly_wage']>=6000].describe()
```

```python
#more_than_6000_2 컬럼에 6000원 이상일 때 A Group이고 그렇지 않으면 B Group 추가
df['more_than_6000_2'] = np.where(df['hourly_wage'] >=6000, 'A Group', 'B Group'  )
df.head()
```

```python
#전체 데이터에서 구 정보와 시급 데이터만을 추출하여 시급 기준 내림차순으로 정렬
df[['area1','hourly_wage']].sort_values(by=['hourly_wage'], ascending=False) 
```

```python
#시급 데이터를 10구간으로 나눈 히스토그램
df.drop(columns=['outlier']).hist(bins=10)
```

```python
#시급 데이터의 박스 플롯
plt.boxplot(df['hourly_wage'])
plt.xlabel(xlabel='hourly_wage')
plt.ylabel(ylabel='wage')
```

```python
sns.boxplot(y="hourly_wage", data=df, palette="Set1" ,width=0.2)
```

```python
#회사별 데이터의 박스 플롯
sns.boxplot(x='name', y='hourly_wage', data=df, width=.2)
```

```python
#지역별 시급 데이터의 박스 플롯
plt.figure(figsize=(18,8))
sns.boxplot(x='area1', y='hourly_wage', data= df, width=.2)
```
