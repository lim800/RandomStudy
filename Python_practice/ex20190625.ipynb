# 주식 시세 가져오기

```python
url = 'http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13' 
code_df = pd.read_html(url,header=0)[0]
type(code_df) 
code_df.head()
```

```python
code_df = code_df[['회사명','종목코드']]
code_df.head(10)
```

```python
# 컬럼 영문화
code_df.columns = ['company','code']
code_df.head()
```

```python
item_name='LG이노텍'
```

```python
def get_url(item_name, code_df):
    code = code_df[code_df['company']==item_name]['code'].iloc[0].astype(str).zfill(6)
    print(code)
    url = 'https://finance.naver.com/item/sise_day.nhn?code={code}'.format(code=code) 
    print("요청 URL = {}".format(url))
    return url

url = get_url(item_name, code_df)
```

```python
url = url+r'&page='
url
```

```python
sise = pd.read_html(url+'{}'.format(1), header=0)[0].dropna()
sise
```

### 최대 페이지 가져오기

```python
import requests
page = requests.get(url+'1')
page
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print( soup.prettify())
```

```python
atag = soup.find_all('a')[len(soup.find_all('a'))-1]
atag = str(atag).split("page=")
last_page = int(atag[1].split(r'">맨뒤')[0])
last_page
```

```python
for i in list(range(2,last_page+1)):
    sise = sise.append(pd.read_html(url+'{}'.format(i), header=0)[0].dropna(), ignore_index=True)
```

```python
sise.sort_values('날짜',ascending=False).head()
```

```python
sise.iloc[:,1:6] = sise.iloc[:,1:6].astype(int)
```

```python
sise['날짜']=pd.to_datetime(sise['날짜'])
```
