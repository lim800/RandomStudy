XML 데이터 작성

```python
note = Element('note')
note.attrib['date'] = '20190604'
to = Element('to')
to.text = '파이썬'
note.append(to)
SubElement(note, 'from').text = '내용'
SubElement(note, 'heading').text = '의미'
SubElement(note, 'body').text = '객체지향 언어의 꽃이다'
SubElement(note, 'tail').text =  '그렇다'

def indent(elem, level=0):
    i = '\n' + level * '    '
    if len(elem):
        if not elem.text or not elem.text.strip():
            elem.text = i + '    '
        if not elem.tail or not elem.tail.strip():
            elem.tail = i
        for elem in elem:
            print('text: ',elem.text, bool(elem.text))
            print('tail: ',elem.tail, bool(elem.tail))
            indent(elem, level+1)
        if not elem.tail or not elem.tail.strip():
            elem.tail = i
    else:
        if level and (not elem.tail or not elem.tail.strip()):
            elem.tail = i
indent(note)
dump(note)
ElementTree(note).write('note.xml', xml_declaration=True, encoding='utf-8')
```

```python
members = ['임꺽정', '양산박', '로빈후드', '루팡']
nations = [['한국','임꺽정'],['중국','양산박'],['영국','로빈후드'],['프랑스','루팡']]
root = etree.Element('group')
w_name = etree.Element('name')
root.append(w_name)
w_name.text = '의적들'
w_name.set('alias', 'companion')

w_members = etree.Element('members')
for member in members:
    w_member = etree.SubElement(w_members, 'member')
    w_member.text = member
root.append(w_members)

w_nations = etree.Element('nations')
for nation in nations:
    w_nation = etree.SubElement(w_nations, 'nation')
    w_nation.text = nation[0]
root.append(w_nations)

# etree.dump(root)
out = etree.tostring(root, pretty_print=True, encoding ='utf-8')
head = "<?xml version='1.0' encoding='utf-8'?>\n"
print(head + out.decode('utf-8'))

etree.ElementTree(root).write('group.xml', pretty_print = True, encoding='utf-8', xml_declaration = True)
```
