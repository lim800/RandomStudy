# TensorFlow

```python
import tensorflow as tf
```

```python
#공부 시간
x_data = [2, 4, 6, 8]
#점수
y_data = [ 81, 93, 91, 97]
```

```python
tf.reset_default_graph()
sess =  tf.Session()
#독립 변수와 종속 변수의 미리 저장하는 공간
X = tf.placeholder(tf.float32, shape=None)
Y = tf.placeholder(tf.float32, shape=None)
#가중치 설정
W = tf.Variable(tf.random_normal([1], name='weight'))
#편향
b = tf.Variable(tf.random_normal([1], name='bias'))
```

```python
#가설 설정 -> 함수를 입력
hypo = W*X + b
#손실 함수
cost = tf.reduce_mean(tf.square(hypo-Y))
#손실 함수 최소와 방법
opti = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = opti.minimize(cost)
print(train)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(4001):
        cost_val, W_val, b_val, train_val = sess.run([cost,W,b,train], feed_dict={X: x_data, Y: y_data})
        if step % 40 == 0:
            print(step, "손실 : " , cost_val, "\t가중치 : ", W_val, "\t편향 : ", b_val)
    print("H(4) 예측 : ", sess.run(hypo, feed_dict={X: [8]}))
```

```python
tf.reset_default_graph()

x1_data = [73, 93, 89, 96, 73]
x2_data = [80, 88, 91, 98, 66]
x3_data = [75, 93, 90, 100, 70]
y_data = [72, 88, 92, 100, 71]

x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random.normal([1]), name='weight1')
w2 = tf.Variable(tf.random.normal([1]), name='weight2')
w3 = tf.Variable(tf.random.normal([1]), name='weight3')

b = tf.Variable(tf.random.normal([1]), name='bias')
```

```python
hypo = x1*w1 + x2 * w2 +x3 * w3 + b
cost = tf.reduce_mean(tf.square(hypo - Y))
opti = tf.train.GradientDescentOptimizer(learning_rate=0.00001)
train = opti.minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        cost_val, hypo_val, _ = sess.run([cost, hypo, train], feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
        
        if step % 10 == 0:
            print(step, "손실: ", cost_val, "\t예측: ", hypo_val)
```
