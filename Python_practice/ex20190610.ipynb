```python
sql = """
drop table if exists fruit_prices;
create table fruit_prices(
fruit_id integer primary key,
name text unique,
price integer
);

insert into fruit_prices (name, price) values('사과', 800);
insert into fruit_prices (name, price) values('오렌지', 780);
insert into fruit_prices (name, price) values('바나나', 430);


"""

cur.executescript(sql) # 여러줄의 명령문 실행
conn.close()

```

```python
conn = sqlite3.connect('./sqlite/concon.db')
cur = conn.cursor()
sql = "select * from fruit_prices"
select = cur.execute(sql) # 한줄의 명령문 실행
print(select.fetchall())
```

```python
conn = sqlite3.connect('./sqlite/concon.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    sql = "select * from fruit_prices"
    s = cur.execute(sql)
#     print(s.fetchall()) -> 한줄에 전체 출력됨
    print(s.fetchmany(2)) # (두개)의 인수 출력
```

```python
conn = sqlite3.connect('./sqlite/concon.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    sql = "select * from fruit_prices"
    cur.execute(sql)
    names = [des[0] for des in cur.description]
    print(names)
```

```python
conn = sqlite3.connect('./sqlite/concon.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    sql = "select * from fruit_prices"
    cur.execute(sql)
    # 람다 함수 사용
    names = list(map(lambda x: x[0] ,cur.description))
    print(names)
```

```python
# execute

conn = sqlite3.connect('./sqlite/concon.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    sql = "insert into fruit_prices(name, price) values(?, ?)"
    data = '호박', 5500
    cur.execute(sql, data)
    sql = "select * from fruit_prices"
    cur.execute(sql)

    fruit_list =cur.fetchall()
    print(fruit_list)
```

```python
# executemany

conn = sqlite3.connect('./sqlite/concon.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    sql = "insert into fruit_prices(name, price) values(?, ?)"
    data = [('망고',7700),('키위',4000),('감',7000)]
    cur.executemany(sql, data)
    sql = "select * from fruit_prices"
    cur.execute(sql)

    fruit_list =cur.fetchall()
    print(fruit_list)   
```

```python
#update

conn = sqlite3.connect('./sqlite/concon.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    sql = "update fruit_prices set name = ?, price = ? where fruit_id = ?"
    data = '청포도',900,4
    cur.execute(sql, data)
    sql = "select * from fruit_prices"
    cur.execute(sql)

    fruit_list =cur.fetchall()
    print(fruit_list)   
```

```python
# delete

conn = sqlite3.connect('./sqlite/dbtest.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    data = [(4,),(5,),(6,)]
    sql = "delete from fruit_prices where fruit_id = ?"
    cur.executemany(sql, data)
    sql = "select * from fruit_prices"
    cur.execute(sql)
    fruit_list = cur.fetchall()
    print(fruit_list)
```

```python
# dictionary 사용

conn = sqlite3.connect('./sqlite/dbtest.db', isolation_level=None) # 자동 커밋 설정
with conn:
    cur = conn.cursor()
    data = {"price1": 300, "price2": 780}
    sql = "select * from fruit_prices where price >= :price1 and price <= :price2"
    cur.execute(sql, data)
    lists = cur.fetchall()
    print(lists)
    
    
```

```python
import csv
conn = sqlite3.connect('./sqlite/csvtodb.db', isolation_level=None)
with conn:
    cur = conn.cursor()
    cfile = open('/Users/lesson6_mac/AtomData/python/sqlite/gtest.csv')
    creader = csv.reader(cfile)
    dlist = list(creader)[1:]
    data = []
    for a in dlist:
        data.append([a[1],a[8],a[9],a[10]])
    sql = "insert into kilo values(?,?,?,?)"
    cur.executemany(sql,data)
    cur.execute("select * from kilo")
    print(cur.fetchall())
```
