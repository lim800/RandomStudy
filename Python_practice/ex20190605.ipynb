## bs4

```python
from bs4 import BeautifulSoup
import urllib.request as ureq
import os.path as pt
```

```python
url = "http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
savename = 'forecat.xml'
if not pt.exists(savename):
    ureq.urlretrieve(url, savename)
xml = open(savename, encoding='utf-8').read()
print(xml)
```

```python
soup = BeautifulSoup(xml, 'lxml-xml')
info = {}

for location in soup.find_all('location'):
    name = location.find('city').string
    weather = location.find('wf').string
    if not weather in info:
        info[weather] = []
    info[weather].append(name)
for weather in info.keys():
    print('날씨: ', weather)
    for name in info[weather]:
        print('-', name)
```

## lxml

```python
from lxml import etree
from xml.etree.ElementTree import Element, dump, SubElement, ElementTree, parse
```

```python
tree = etree.parse('forecat.xml')
root = tree.getroot()
# print(root)
dump(root)
```

```python
weather = {}
for loc in root.iter('location'):
    for j in loc:
        if j.tag == 'city':
            city_name = j.text
            weather[city_name] = []
        if j.tag == 'data':
            for i in j:
                if i.tag == 'wf':
                    weather_stat = i.text
                    weather[city_name].append(weather_stat)
                        

print(weather)
                
```

```python
wth = []
for a, b in weather.items():
    wth.append(b[0])

s_wth = list(set(wth))
print(s_wth)
```

```python
for stat in s_wth:
    print('날씨: ', stat)
    for a, b in weather.items():
        if b[1] == stat:
            print(a)
    
```

```python
from lxml import etree
from xml.etree.ElementTree import Element, dump, SubElement, ElementTree, parse
tree = etree.parse('forecat.xml')
root = tree.getroot()

weather = {}
for loc in root.iter('location'):
    for j in loc:
        if j.tag == 'city':
            city_name = j.text
            weather[city_name] = []
        if j.tag == 'data':
            for i in j:
                if i.tag == 'wf':
                    weather_stat = i.text
                    weather[city_name].append(weather_stat)
wth = []
for a, b in weather.items():
    wth.append(b[0])
s_wth = list(set(wth))

# 결과 출력
for stat in s_wth:
    print('날씨: ', stat)
    for a, b in weather.items():
        if b[3] == stat:
            print(a)
```

```python
import urllib.request as ureq
import os.path
import json
url = "https://api.github.com/repositories" 
savename = 'repository.json'
# 저장할 파일 존재 확인
if not os.path.exists(savename):
    ureq.urlretrieve(url, savename)
data = open(savename, encoding="utf-8").read( ) 
items = json.loads(data)
# 아이템을 반복하여 저장소 이름과 소유자 확인
# print(items)

for item in items:
    print('저장소 이름은 {}이며 소유자는 {}입니다.'.format(item['name'], item['owner']['login']))
```

```python
with open(savename, encoding="utf-8") as f:
    git_n = json.load(f)
for item in git_n:
    print('저장소 이름은 {}이며 소유자는 {}입니다.'.format(item['name'], item['owner']['login']))
```
