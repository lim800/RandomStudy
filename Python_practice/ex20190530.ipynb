## RSS

```python
import urllib.request as ureq
import urllib.parse as upa

print('지역의 선택은 전국, 수도권, 강원, 충북, 충남, 전남, 전북, 경북, 경남, 제주 중 한 곳을 선택하세요.')

# 조건이 맞을 때까지 반복한다.
while True:
    region_name = input('지역 선택: ')
    region_number = { "전국": 108, "수도권": 109, "강원": 105, "충북": 131, "충남": 133,\
                     "전북": 146, "전남": 156, "경북": 143, "경남": 159, "제주": 184 } # 키로 지역 존재를 확인하지 않는다.
    
    if region_name not in region_number.keys( ): 
        print('지역 이름을 잘못 입력하셨습니다.') 
        continue
    
    # 키로 지역 존재를 확인한다.
    if region_name in region_number.keys( ):
        select_number = region_number[region_name]
        rss = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp" 
        query = {"stnId": select_number}
        params = upa.urlencode(query)
        url = rss + '?' + params
        print('url=', url)
        data = ureq.urlopen(url).read( )
        text = data.decode('utf-8')
        print(text)
        break
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(text, 'html.parser')
print(soup)
```

```python
title = soup.find('title').string
time = soup.find('tm').string
weather = soup.find('wf').string

print('title: ',title)
print('time: ', time)
print('weather: ', weather)
```

## API

```python
import urllib.request as ureq 
from bs4 import BeautifulSoup

url = "http://ip-api.com/xml" 
request_test = ureq.urlopen(url) 
data = request_test.read( )

# 바이너리 데이터를 문자열로 변환한다.
text = data.decode('utf-8') 

print(text)
```

```python
soup = BeautifulSoup(text, 'html.parser')

country = soup.find('country').string
city = soup.find('city').string
ipp = soup.findAll('query')[1].string

print('country: ', country)
print('city: ', city)
print('ip: ', ipp)
```
